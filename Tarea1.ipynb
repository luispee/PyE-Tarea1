## Parte 1
### 3) Se define calcular_X() que recibe una lista de ceros y unos representando las tiradas de moneda.

```python
def calcular_X(resultados):
    """
    resultados: lista con valores 0 o 1
    """
    X = 0
    for i, Mi in enumerate(resultados, start=1):
        X += Mi / (2 ** i)
    return X
```

```python
ejemplo = [1, 0, 1, 1, 0, 0, 1, 0, 1, 1]
X = calcular_X(ejemplo)
print(f"El valor de X para el ejemplo es: {X}")
```

### El metodo product de itertools genera todas las combinaciones posibles de 0 (cara) y 1 (número) para 3 lanzamientos de moneda

```python
t = product([0, 1], repeat=3) 
print(list(t))
```

### 4) Genera todas las combinaciones posibles de "cara" y "número" para 10 lanzamientos, calculo x y lo agrega al recorrido, se retorna la cantidad de elementos


```python
recorrido = set()
for resultados in product([0, 1], repeat=10):
    X = calcular_X(resultados)
    recorrido.add(X)
elementos = len(recorrido)
print(f"Elementos distintos en el recorrido: {elementos}")        
```

### 6) Definimos la funcion analizar_X para calcular esperanza y varianza, su parámetro es la cantidad de monedas, que en la próxima parte vamos a hacer para 5 valores distintos.

```python

def analizar_X(i):
    recorrido_x = set()
    for resultados in product([0, 1], repeat=i):
        X = calcular_X(resultados)
        recorrido_x.add(X)
    recorrido_x = list(recorrido_x)
    media = np.mean(recorrido_x)
    varianza = np.var(recorrido_x)
    return media, varianza
```

### 8) Comprobamos que la esperanza y la varianza se aproximan hacia la uniforme continua, cuantas más monedas se utilizan más se aproxima al valor teórico.

```python
cantidades = [10, 12, 14, 16, 18, 20]

print(f"{'Monedas':<10} | {'Esperanza':<12} | {'Varianza':<12}")
print("-" * 40)

for k in cantidades:
    E, Var = analizar_X(k)
    print(f"{k:<10} | {E:<12.5f} | {Var:<12.8f}")
```

### 2) Para la simulación de n experimentos definimos N_simulaciones y utilizamos calcular_X sobre la lista de monedas (caras y números) aleatorios

```python
def N_simulaciones(veces,monedas):
    simulacion = []
    for i in range(0,veces):
        simulacion.append(calcular_X(simular_X(monedas)))   
    return simulacion
```

### Loop de simulaciones, se guardan los valores de x sobre las cantidades n de experimentos de 10 monedas

```python
n = [10**3, 10**4, 10**5, 10**6]
simulaciones = []

for k in n:
    simulaciones.append(N_simulaciones(k,10))
```

### 3) Definimos la funcion para calcular la frecuencia relativa del evento { 2/7 ≤ X ≤ 6/7 }

```python
def frecuencia_Relativa(lista):
    x = 0
    for i in lista:
        if i >= 2/7 and i <= 6/7:
            x += 1
    return x / len(lista)
```

### Utilizamos las simulaciones para calcular la frecuencia relativa

```python
print(f"{'Simulacion':<10} | {'Frecuencia relativa':<12}")
print("-" * 40)

for k in range(0,4):
    F = frecuencia_Relativa(simulaciones[k])
    print(f"{n[k]:<10} | {F:<12.5} ")
```

### 1) Simulamos 2 tiradas independientes de n cantidad de monedas equilibradas 

```python
def simular_XY(monedas):
    resultados1 = np.random.randint(0, 2, size=monedas)
    resultados2 = np.random.randint(0, 2, size=monedas)
    return resultados1, resultados2
```

### Definimos una función para calcular los valores de XY utilizando la funcion calcular_X definida en la parte 1

```python
def N_Simulaciones_XY(veces,monedas):
    simulacion = []
    for i in range(0,veces):
        (a,b) = simular_XY(monedas)
        simulacion.append((calcular_X(a), calcular_X(b)))
    return simulacion
```

### Ejecutamos las simulaciones para 10<sup>3</sup>, 10<sup>4</sup>, 10<sup>5</sup>, 10<sup>6</sup>

```python
n = [10**3, 10**4, 10**5, 10**6]
simulaciones_XY = []
for k in n:
    simulaciones_XY.append(N_Simulaciones_XY(k,10))
```

### Definimos la funcion para calcular la frecuencia relativa del evento {X<sup>2</sup> + Y<sup>2</sup> <= 1 }

```python
def frecuencia_Relativa_XY(lista):
    x = 0
    for i in lista:
        cuadrado = (i[0]**2 + i[1]**2)
        if cuadrado <= 1:
            x += 1
    return x / len(lista)
```

### Ejecutamos el cálculo de las frecuencias relativas sobre la lista de simulaciones de XY, la múltiplicación  se realiza en el print y se aprecia en la aproximación a π

```python
print(f"{'Simulacion':<10} | {'Frecuencia relativa':<12} | {'aproximacion a π':<12} ")
print("-" * 40)

for k in range(0,4):
    F = frecuencia_Relativa_XY(simulaciones_XY[k])
    print(f"{n[k]:<10} | {F:<12.5} | {F*4:<12.5} ")
```
