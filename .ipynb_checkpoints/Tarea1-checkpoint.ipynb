```python
def calcular_X(resultados):
    """
    resultados: lista o array de longitud 10 con valores 0 o 1
    """
    X = 0
    for i, Mi in enumerate(resultados, start=1):
        X += Mi / (2 ** i)
    return X
```

```python
resultados = [1, 0, 1, 1, 0, 0, 1, 0, 1, 1]
X = calcular_X(resultados)
print(X)
```
